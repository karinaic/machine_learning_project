
* La red neuronal:  Su unidad de procesamiento es la neurona, y podemos juntar muchas neuronas en diferentes capas para conseguir esta codificacion de informacion m√°s compleja y que pueda procesar esta informaci√≥n.

* Info: casi todas las formulas han resultado ser posibles de vectorizar, asi se consigue un mejor rendimiento a la hora de conseguir resultados; dentro de una capa estamos realizando las mismas operaciones para todas las neuronas, esto quiere decir si nosotros consideramos a la red nueronal como si fuera la unidad de procesamiento de la neurona, podemos pensar que cada capa es un modulo y el modulo principal es la capa y cada una se diferencia de la otra; esto nos permite definir en una red neuronal una clase u objeto que se refiera a una neurona

#### Clase de la capa de la red
